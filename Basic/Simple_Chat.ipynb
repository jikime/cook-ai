openAI API key 가져와서 API 호출

```python
import os
from openai import OpenAI

api_key = os.getenv("OPENAI_API_KEY")

client = OpenAI(
  api_key=api_key,
)
```

```python
query = "안녕하세요!"

chat_completion = client.chat.completions.create(
    messages=[
      {
        "role": "system", 
        "content": "You are a helpful assistant."
      },
      {
          "role": "user",
          "content": query,
      }
    ],
    max_tokens=300,
    temperature=0,
    model="gpt-3.5-turbo",
)
```

```python
chat_completion.choices[0].message.content
```

템플릿을 활용해 프롬프트 만들기

```python
from string import Template

prompt_template = Template(f"""
다음 사용자의 요청에 답변해주세요.
사용자: {query}"
답변:
""")

print(prompt_template.substitute(query=query))
```

```python
system_prompt = """
당신은 상담을 해주는 인공지능 에이전트입니다. 친절하고 자세하게 사용자와 대화를 나눠주세요. 필요하면 이모티콘을 사용해도 좋습니다.
"""
```

```python
def QnA(query):
  prompt = prompt_template.substitute(query=query)
  chat_completion = client.chat.completions.create(
      messages=[
          {
            "role": "system",
            "content": system_prompt,
          },
          {
              "role": "user",
              "content": prompt,
          }
      ],
      model="gpt-3.5-turbo",
      max_tokens=300,
      temperature=0,
  )
  return chat_completion.choices[0].message.content

QnA("안녕하세요! 만나서 반가와요!")
```
