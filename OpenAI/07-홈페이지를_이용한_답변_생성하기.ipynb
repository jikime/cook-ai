### 홈페이지 기반 답변 생성하기

##### 1. 홈페이지에서 정보 가져오기

```python
import requests
from bs4 import BeautifulSoup

url = 'https://fastcampus.co.kr/data_online_aicc?utm_source=facebook&utm_medium=viral&utm_campaign=prd%5E240402%5E236246&utm_content=teacher%5Ekim%5E236246'

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup)

else :
    print(response.status_code)
```

```python
data = soup.get_text()
print(data)
```

##### 2. GPT API를 이용하여 중요한 정보 구조화 하기

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from openai import OpenAI

model_name="gpt-4o-mini"

client = OpenAI()

def summarize_html(prompt):
  chat_completion = client.chat.completions.create(
      messages=[
          {
              "role": "user",
              "content": prompt,
          }
      ],
      model=model_name,
  )
  return chat_completion.choices[0].message.content
```

```python
def promt_generation(raw_txt):
    promt = f"""
    TXT는 html문서에서 추출된 텍스트입니다.
    TXT를 보고 정보를 구조화 하여서 json 파일을 만들어주세요.
    포함되어야 하는 정보는 코스 제목, 강사진, 강의 내용, 추천 대상입니다.

    TXT: {raw_txt}

    JSON:
    """
    return promt

prompt = promt_generation(data)


structured_data = summarize_html(prompt)
```

```python
print(structured_data)
```

##### 4. 웹 사이트 정보기반 답변 생성


```python
from string import Template

query = "안녕하세요. 이 강의에서 배울 수 있는 것들을 알려주세요"

system_prompt = Template("""
너는 $course_title 관련 답변을 해주는 인공지능 챗봇이야.
주어진 강의 관련 정보를 활용하여서 사용자에게 친절한 답변을 제공해줘.

강의 관련 정보
$informations

""")

prompt = Template("""
다음 사용자의 요청에 대한 답변을 해줘

요청: $query
""")

print(system_prompt.substitute(course_title="25개 실습으로 끝내는 AI 고객센터 서비스 구축", informations=structured_data))
print(prompt.substitute(query=query))
```

```python

def create_answer(query):
  chat_completion = client.chat.completions.create(
      messages=[
          {
              "role": "system",
              "content": system_prompt.substitute(course_title="25개 실습으로 끝내는 AI 고객센터 서비스 구축", informations=structured_data),
          },
          {
              "role": "user",
              "content": prompt.substitute(query=query),
          }
      ],
      model=model_name,
  )

  return chat_completion.choices[0].message.content


query_list = [
    '강사진에 대한 설명을 부탁해',
    '이 강의에서 배울 수 있는 것들을 알려주세요'
]

for q in query_list:
  print(f"Q: {q}")
  print(f"A: {create_answer(q)}")
  print("\n===============================\n")
```
