### 1. OpenAI API를 활용해 GPT와 대화하기

기본 쿼리 입력 및 반환

```python
client = OpenAI()
query = "안녕하세요!"
```

```python
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": query,
        }
    ],
    model="gpt-4o-mini",
  )
chat_completion
```

```python
chat_completion.choices[0].message.content
```

템플릿을 활용해 프롬프트 만들기

```python
from string import Template

prompt_template = Template("""
다음 사용자의 요청에 답변해줘.
사용자: "$query"
답변:
""")

print(prompt_template.substitute(query=query))
```

```python
system_prompt = """
너는 상담을 해주는 인공지능 에이전트야. 친절하고 자세하게 사용자와 대화를 이어나가야 해. 항상 사용자가 궁금하거나 불편한 점이 없는지 확인해줘.

"""
```

```python
def answer(query):
  prompt = prompt_template.substitute(query=query)
  chat_completion = client.chat.completions.create(
      messages=[
          {
            "role": "system",
            "content": system_prompt,
          },
          {
            "role": "user",
            "content": prompt,
          }
      ],
      model="gpt-3.5-turbo",
  )
  return chat_completion.choices[0].message.content

answer("안녕하세요! 처음 뵙겠습니다.")
```

```python
answer("AI는 무엇인가요? 요약해주세요.")
```
