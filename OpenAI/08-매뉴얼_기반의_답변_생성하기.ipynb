### 매뉴얼 기반 답변 생성하기

In-context Learning (ICL)을 통해 외부 정보를 통합할 수 있습니다.

##### 1. 매뉴얼 불러오기

```python
path = '../Assets/data/이용약관및규정.txt'

policy = ""
with open(path, "r") as f:
    policy = f.read()

print(policy)
```

##### 2. 프롬프트 만들기

```python
from string import Template

query = "안녕하세요. 강의를 환불하고 싶어요. 어떻게 해야 하나요?"

system_prompt = Template("""
너는 온라인 강의에 대해 환불 및 결제를 담당해 주는 인공지능 챗봇이야.
이용약관을 참고하여서 사용자의 질문에 친절하게 답변해줘.

이용약관
$terms
""")

prompt = Template("""
다음 사용자의 요청에 대한 답변을 해줘

요청: $query
""")

print(system_prompt.substitute(terms=policy))
print(prompt.substitute(query=query))
```

##### 3. GPT api를 활용하여 답변생성하기

```python
from openai import OpenAI

client = OpenAI()

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": system_prompt.substitute(terms=policy),
        },
        {
            "role": "user",
            "content": prompt.substitute(query=query),
        }
    ],
    model="gpt-4o-mini",
)

print(chat_completion.choices[0].message.content)
```

##### 4. 함수로 정의하여 다양한 답변에 대해 대답하기

```python

def create_answer(query):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": system_prompt.substitute(terms=policy),
            },
            {
                "role": "user",
                "content": prompt.substitute(query=query),
            }
        ],
        model="gpt-40-mini",
        temperature = 0.5
    )

    return chat_completion.choices[0].message.content


query_list = [
    '파이썬을 활용한 웹 데이터 수집 강의에서는 무엇을 배울 수 있나요?',
    '강의를 결제하고 싶은데, 어떤 수단을 사용할 수 있나요?'
]

for q in query_list:
    print(f"Q: {q}")
    print(f"A: {create_answer(q)}")
    print("\n===============================\n")
```
